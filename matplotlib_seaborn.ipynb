```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# from matplotlib.legend import Legend
num = np.arange(1,11)
sq = num**2
cub = num**3

plt.figure(figsize=(3,3))

plt.title('linechart')
#fontsize, color

plt.plot(num,sq)
# marker, markersize, markeredgecolor, markerfacecolor, linewidth, linestyle, color

plt.xlabel('numbers')
plt.ylabel('squares')

#fontsize, color, fontweight = 500/'bold'

plt.xticks(range(1,11))
# rot

plt.show()
```

```python
plt.figure(figsize=(10,3))

plt.title('linechart')

plt.plot(num,cub,marker='.')
plt.xlabel('numbers')
plt.ylabel('cubes')
plt.xticks(range(1,11))

#annotate - naming each point in the chart

for i in range(len(num)):
  plt.annotate(f'{num[i]},{cub[i]}',(num[i],cub[i]),textcoords="offset points",xytext=(-10,5))
  #first part is for text display, second part is for where to display the coordinates

plt.show()
```

```python
#figure():
#i)to adjust size of the plot.
#ii)to draw a grid inside another grid

#syntax: plt.figure(figsize=(w,h))
plt.figure(figsize=(10,3))


plt.title('linechart of cubes and squares')
plt.plot(num,sq,label= 'sqr')
plt.plot(num,cub,label = 'cube')
plt.xlabel('numbers')
plt.ylabel('squares and cubes')
plt.xticks(range(1,11))

plt.legend(title='to the power')
# shadow =True, facecolor, loc = best
#loc-->best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

plt.show()
```

```python
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/sales_data_boolean_filtering.csv")
# df.head()
```

```python
region=df.groupby('Region')['TotalSales'].sum()

region=pd.DataFrame(region)
region.reset_index(inplace=True)
r=region.sort_values(by='TotalSales')
r

plt.plot(r['Region'],r['TotalSales'],marker='o',markerfacecolor='yellow',markeredgecolor='yellow',markersize=10,linewidth=3,linestyle='dashed')
plt.xlabel('Region',color='purple',fontsize=15)
plt.ylabel('TotalSales',color='purple',fontsize=15)
plt.title('TotalSales by Region',color='red',fontsize=18)
plt.grid(True,linestyle='dashed',color='red')
```

```python
region=df.groupby('Product')['TotalSales'].sum()

region=pd.DataFrame(region)
region.reset_index(inplace=True)
r=region.sort_values(by='TotalSales')
r

plt.plot(r['Product'],r['TotalSales'],marker='o',markerfacecolor='green',markeredgecolor='green',markersize=7,linewidth=3,linestyle='solid')
plt.xlabel('Product',color='purple',fontsize=15)
plt.ylabel('TotalSales',color='purple',fontsize=15)
plt.title('TotalSales by Product',color='red',fontsize=18)
#plt.grid(True,linestyle='dashed',color='red')
plt.xticks(rotation=90,color='red')


plt.show()
```

```python
#subplot():it is used to create multiple grids
#syntax:plt.subplot(no.of rows,no.of cols,position)
#note:it creates imaginary grids

plt.figure(figsize = (15,15))

plt.subplot(3,3,1)
plt.subplot(3,3,2)
plt.subplot(3,3,3)
plt.subplot(3,3,4)
plt.subplot(3,3,5)
plt.plot(num,sq,label='sqr',marker='o',markersize=8,markerfacecolor='pink',markeredgecolor='pink',linewidth=2,linestyle='solid')
#xlabel-->it is used give name to x axis-->syntax:plt.xlabel('name')
#ylabel-->it is used give name to y axis-->syntax:plt.ylabel('name')
#title():syntax:plt.title('name')

plt.xlabel('values from 1 to 10',color='purple')
plt.ylabel('squares',color='blue')
plt.title('SQUARE',color='red',fontsize=15)

#annotate():it is used to represent data values for data points
for i in range(len(num)):
  plt.annotate(f'({num[i]},{sq[i]})',(num[i],sq[i]))


plt.subplot(3,3,6)
plt.subplot(3,3,7)
plt.plot(r['Product'],r['TotalSales'])
plt.xticks(rotation=90)
plt.subplot(3,3,8)
plt.subplot(3,3,9)
plt.plot(num,cub,label='cube',marker='o',markersize=8,markerfacecolor='red',markeredgecolor='red',linewidth=2,linestyle='solid')
for i in range(len(num)):
  plt.annotate(f'({num[i]},{cub[i]})',(num[i],cub[i]),color='purple',ha='right')
  #ha-->center', 'right', 'left'

plt.xlabel('values from 1 to 10',color='purple')
plt.ylabel('cube of x',color='blue')
plt.title('CUBE',color='red',fontsize=15)

plt.tight_layout()
```

```python
# A pair plot shows pairwise relationships between numerical variables in a dataset all at once.
# Every column vs every other column + each columns distribution

# Diagonal
# Shows distribution

# Default  histogram + KDE

# Upper & Lower Triangle
# Shows scatter plots
# Relationship between variables

#syntax:sns.pairplot(df)

sns.pairplot(df)
```

```python
# A heat map is a 2-D graphical representation of data where numerical values are represented using colors.
# In data analysis, heat maps are most commonly used to visualize correlation matrices.

# Darker / brighter color = higher or lower value (based on colormap)

# Why Heat Maps Are Used
# Heat maps help to:
# Quickly identify relationships
# Detect strong / weak correlations
# Spot patterns & trends
# Reduce complex tables into visual form

# Colormap (cmap)
# Defines color scheme
# Example: coolwarm, viridis, magma

# Annotation (annot)
# Displays numeric value inside each cell
# True / False

# Square matrix showing pairwise correlation
# Values range from 1 to +1

#i) when corr value is equal to +1-->highly correlated(directly)
#iI) when corr value is equal to -1-->negativly  correlated(inversely)

corr_data=df.iloc[:,1:].corr()
sns.heatmap(corr_data,annot=True,cmap='magma')
```

```python
# A pivot table is a data summarization tool that:
# Groups data
# Aggregates values
# Converts long-format data into matrix form
# This matrix is perfect input for a heat map.

# pd.pivot_table(
#     data,
#     values,
#     index,
#     aggfunc
# )

# df.pivot_table(values=['fixed acidity','citric acid','pH','alcohol','chlorides'],index='quality',aggfunc=['mean'])
mean=pd.pivot_table(df,values=['fixed acidity','citric acid','pH','alcohol','chlorides'],index='quality',aggfunc=['mean'])

sns.heatmap(mean,annot=True,cmap='magma')

# Index
# Becomes Y-axis

# Columns
# Becomes X-axis

# Values
# Become cell color

# Aggregation (aggfunc)
# How values are summarized
# (sum, mean, count, max)
```

```python
# | Aspect      | Correlation Heat Map | Pivot Heat Map     |
# | ----------- | -------------------- | ------------------ |
# | Source      | `df.corr()`          | `pd.pivot_table()` |
# | Aggregation | Statistical          | User-defined       |
# | Use         | Feature relationship | Business insights  |
```
