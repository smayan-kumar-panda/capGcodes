# boolean Filtering

```python
# boolean filtering(true or false output)

data['age']>18
data[((data['age']>18) & (data['sex']=='male')) | (data['fare']>500) ]  # gives only the true statements from this table

```

# Grouping

```python
# grouping

# grouping is done on categorical columns
data['sex'].unique()

females=data.groupby('sex').get_group('female')   # gives output as dataframes so we can perform operations on the output

embarked_S_from_females= females.groupby('embarked').get_group('S')
```

```python
embarked_C_from_females= females.groupby('embarked').get_group('C')
embarked_C_from_females
```

# Statistical functions

```python
np.mean(embarked_C_from_females['age'])
```

```python
fare_gt_400= data[data['fare']>400]   # boolean filtering
fare_gt_400.groupby('sex').get_group('female')  # grouping on the output of boolean filtering
```

# aggregate functions

```python
data['fare'].max()
data['fare'].min()
data['fare'].mean()
data['fare'].median()
data['fare'].std()
data['fare'].var()
```

```python
data['survived'].unique()
died=data.groupby('survived').get_group(0)
female_died=died.groupby('sex').get_group('female')
female_died

female_died[female_died['age']<18]
```

```python
data.groupby(['sex','survived']).get_group(('female',1))  # multiple grouping

# count() only counts the not null values
```

```python
# handling missing values

# we can drop the entire row where null/nan null is there

data.dropna()

# fill na--> filling nan/null with something else

filling_cabin= data['cabin'].fillna('unknown')
filling_cabin.unique()

data['cabin']=filling_cabin  # changing the original dataset
data['cabin'].unique()

# fill na--> for integers we can fill with 0,mean,mode

# filling with backward fill and forward fill

data['cabin'].fillna(method='ffill').count()
data['cabin'].fillna(method='bfill').count()

# duplicates

data.duplicated().value_counts()


# conversion of a column's datatype into another datatype
data['age']=data['age'].fillna(0)
data['age'].astype(int)
```
