```python
import pandas as pd
import numpy as np
```

```python
dept_data = {'deptno':[1045,1046,1047,1048,1049],
             'deptname':['sales','support','marketing','IT','accounts'],
             'maxsal':[500000,450000,400000,800000,1000000]}
dept = pd.DataFrame(dept_data)
dept
```

```python
emp1_data = {'empid':[111,112,113,114,115,116,117,118,119,121,122,123,124],
             'name':['SRK','salman','HRX','allu','amir','Khiladi','chiranjeev','ram','sonu','rajpal','anil','kapil','carry'],
             'deptno':[1045,1047,1046,1046,1045,1049,1048,1045,1047,1048,1049,1046,1048],
             'salary':[430000,200000,425000,375000,400000,850000,600000,465000,240000,550000,995000,320000,750000]}

emp1 = pd.DataFrame(emp1_data)
emp1
```

```python
emp2_data = {'empid':[111,112,113,114,115,116,117,118,119,121,122,123,124],
             'name':['SRK','salman','HRX','allu','amir','Khiladi','chiranjeev','ram','sonu','rajpal','anil','kapil','carry'],
             'deptno':[1045,1047,1054,1046,1045,1049,1050,1045,1047,1048,1049,1005,10049],
             'salary':[430000,200000,425000,375000,400000,850000,600000,465000,240000,550000,995000,320000,750000]}

emp2 = pd.DataFrame(emp2_data)
emp2
```

# Inner **Join**

```python
pd.merge(dept,emp1,on='deptno',how='inner')
```

**What it does**


*   Keeps only matching deptno in both tables.
*   Since all emp1 deptno exist in dept  no row loss.

**Result**



*   13 rows (same as emp1)
*   Every employee matched to department
*   No NaN

**Key Point**

*   INNER JOIN returns intersection of keys.





```python
pd.merge(dept,emp2,on='deptno',how='inner')
```

**What it does**


*   Only matching deptnos kept
*   Employees with invalid/not-matching deptno removed

**Rows Lost**


*   empid 113,empid 117,empid 123,empid 124

**Key Point**

*   INNER JOIN removes unmatched rows from both sides





# Left **Join**

```python
pd.merge(dept,emp1,on='deptno',how='left')
```

**What it does**


*   Keeps all rows from dept
*   Adds employees where match exists

**Result**



*   Departments with multiple employees  repeated
*   No missing rows because all depts have employees

**Key Point**

*   LEFT JOIN keeps all left table rows.





```python
pd.merge(dept,emp2,on='deptno',how='left')
```


**Result**
*   All departments kept
*   Only matching employees shown
*   Invalid dept employees ignored

**Key Point**

*   LEFT JOIN protects left table (dept).





# Right **Join**

```python
pd.merge(dept,emp1,on='deptno',how='right')
```

**What it does**

*   Keeps all rows from emp1
*   Adds dept info

**Result**

*   Same as INNER here (because all deptnos match)

**Key Point**

*   RIGHT JOIN keeps all right table rows.





```python
pd.merge(dept,emp2,on='deptno',how='right')
```

**Result**

*   All emp2 rows kept
*   Invalid dept employees show: NaN
---
deptname = NaN; maxsal   = NaN
---
empid 113  deptno 1054  NaN dept

**Key Point**

*   RIGHT JOIN exposes data quality problems.


#Outer **Join**

```python
pd.merge(dept,emp1,on='deptno',how='outer')
```

**What it does**

*   Keeps all rows from both
*   Since all match  same as inner.shape, row order can varry sometimes.





```python
pd.merge(dept,emp2,on='deptno',how='outer')
```

**Result**

*   All dept rows
*   All emp2 rows
*   Invalid dept rows show NaN in dept columns

**Key Point**

*   OUTER JOIN gives full audit view.





**Real-World Interpretation**
---

JoinType----- Business Meaning
---
INNER--------Only valid employee-department relationships
---
LEFT--------Show all departments even if no employees
---
RIGHT--------Show all employees even if invalid dept
---
OUTER--------Full reconciliation / audit
