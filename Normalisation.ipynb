```python
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
df1=pd.read_csv('diabetes (1).csv')
df1
```

```python
# L2 normalization (-1 to 1)

num=Normalizer()
data=num.fit_transform(df1)
df=pd.DataFrame(data,columns=df1.columns)
df
```

```python
# range of standard scaler is no limit both negative and positive values

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
df_scaled
```
