```python
# xgboost classifier and regressor
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris,fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error,mean_absolute_error,f1_score,r2_score
from xgboost import XGBClassifier, XGBRegressor
```

```python
data=load_iris()
data
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df['target']=data.target
df.head()
```

```python
x=df.drop('target',axis=1)
y=df['target']
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=XGBClassifier(n_estimators=100,max_depth=5,learning_rate=0.1,random_state=42)
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
train_eff=model.score(xtrain,ytrain)
test_eff=model.score(xtest,ytest)
print('train_eff- ',train_eff)
print('test_eff- ',test_eff)
```
